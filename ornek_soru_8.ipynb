```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Veri seti
data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0]])

# Ortalamaların çıkarılması
mean = np.mean(data, axis=0)
data_centered = data - mean

# PCA uygulama
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_centered)

# Sonuçların görselleştirilmesi
plt.scatter(pca_result[:, 0], pca_result[:, 1], color='blue', label='PCA Sonuçları')
plt.xlabel('Birinci Temel Bileşen')
plt.ylabel('İkinci Temel Bileşen')
plt.title('PCA Sonuçları')
plt.legend()
plt.grid(True)
plt.show()
```
